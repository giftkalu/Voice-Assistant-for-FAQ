```python

import torch
import transformers
import whisper
#from elevenlabs.client import ElevenLabs
from sentence_transformers import SentenceTransformer
import numpy as np
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
import json
import spacy.cli
spacy.cli.download("en_core_web_sm")
import spacy
import io
from gtts import gTTS
```

```python
whisper_model = whisper.load_model("base")  # Options: tiny, base, small, medium, large
embedding_model = SentenceTransformer('all-MiniLM-L6-v2')  # Lightweight and fast
```

```python
faq_df=pd.DataFrame(pd.read_csv("500faq.csv"))
faq_df.head()
```

```python
nlp = spacy.load("en_core_web_sm")

# Combined preprocessing: lowercase, remove punctuation, lemmatize
def preprocess_question(q):
    doc = nlp(q.lower())
    tokens = [
        token.lemma_ for token in doc
        if not token.is_punct and not token.is_space
    ]
    return " ".join(tokens)
faq_df['clean_question'] = faq_df['question'].apply(preprocess_question)
```

```python
questions = faq_df['clean_question'].tolist()
question_embeddings = embedding_model.encode(questions)
print(f"Generated embeddings with shape: {question_embeddings.shape}")
```

**FAQ NORMAL**

```python
def find_best_faq_match(user_question, top_k=1, threshold=0.7):
    """
    Find the best matching FAQ for a user question using semantic similarity.
    Tries 3 strategies:
    1. Original question
    2. Append "cowrywise"
    3. Replace "you"/"u" with "cowrywise"
    """

    def get_best_match(query):
        user_embedding = embedding_model.encode([query])
        similarities = cosine_similarity(user_embedding, question_embeddings)[0]

        best_idx = np.argmax(similarities)
        best_similarity = similarities[best_idx]

        return best_idx, best_similarity

    # --- Strategy 1: Original question
    best_idx, best_similarity = get_best_match(user_question)
    if best_similarity >= threshold:
        return {
            'question': faq_df.iloc[best_idx]['question'],
            'answer': faq_df.iloc[best_idx]['answer'],
            'similarity': float(best_similarity),
            'index': best_idx
        }

    # --- Strategy 2: Append "cowrywise"
    best_idx, best_similarity = get_best_match(user_question + " cowrywise")
    if best_similarity >= threshold:
        return {
            'question': faq_df.iloc[best_idx]['question'],
            'answer': faq_df.iloc[best_idx]['answer'],
            'similarity': float(best_similarity),
            'index': best_idx
        }

    # --- Strategy 3: Replace "you" or "u" with "cowrywise"
    replaced_question = (
        user_question.replace(" you ", " cowrywise ")
        .replace(" u ", " cowrywise ")
        .replace("you", "cowrywise")  # catch edge cases
        .replace("u", "cowrywise")
    )
    if replaced_question != user_question:  # only retry if something actually changed
        best_idx, best_similarity = get_best_match(replaced_question)
        if best_similarity >= threshold:
            return {
                'question': faq_df.iloc[best_idx]['question'],
                'answer': faq_df.iloc[best_idx]['answer'],
                'similarity': float(best_similarity),
                'index': best_idx
            }
    if best_similarity >= 0.4:
        return {
            'question': faq_df.iloc[best_idx]['question'],
            'answer': "This seems related to your question: " + faq_df.iloc[best_idx]['answer'] + "Hope this helps",
            'similarity': float(best_similarity),
            'index': best_idx
        }

    # --- Fallback: return best guess but with unsure answer
    return {
        'question': faq_df.iloc[best_idx]['question'],
        'similarity': float(best_similarity),
        'answer': "I'm not sure how to answer that right now. Try rephrasing your question."
    }
```

```python
test_questions = [
    "What does cowrywise do?",
    "How do I see my earnings?",
    "which of mutual funds or savings have more interest?",
    "Can I link my Opay bank account?",
    "can I earn two hundred thousand naira with ten thousand naira",
    "CAN I INVEST IN STOCKS ON COWRYWISE"
]

processed_test_questions = [preprocess_question(q) for q in test_questions]

for original_q, processed_q in zip(test_questions, processed_test_questions):
    match = find_best_faq_match(processed_q)  # now returns a dict

    print(f"User Q: {original_q}")
    print(f"Best Match: {match['question']}")
    print(f"Similarity: {match['similarity']:.3f}")
    print(f"Answer: {match['answer'][:150]}...")
    print("-" * 30)
```

```python
import whisper
import os
import tempfile
import librosa
import soundfile as sf
import numpy as np
from pathlib import Path
import time
from pydub import AudioSegment, effects
import noisereduce as nr
import warnings
## import sounddevice as sd
warnings.filterwarnings('ignore')
```

```python
def preprocess_and_transcribe(audio_path):
    # Step 1: Load
    y, sr = librosa.load(audio_path, sr=None)

    # Step 2: Reduce noise
    reduced = nr.reduce_noise(y=y, sr=sr)
    sf.write("cleaned.wav", reduced, sr)

    # Step 3: Normalize
    sound = AudioSegment.from_wav("cleaned.wav")
    normalized = effects.normalize(sound)
    normalized.export("normalized.wav", format="wav")

    # Step 4: Transcribe with bias
    model = whisper.load_model("base")
    result = model.transcribe("normalized.wav", language="en", initial_prompt="balance, account, payment, transaction\
    ,mutual funds, savings, opay, direct , interest,  customer, earnings, Cowrywise, fintech, plan, rate, returns, bank, naira")

    transcript = result["text"]
    return transcript
```
